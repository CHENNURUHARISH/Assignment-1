```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#scatter graph
df=pd.read_csv('covid.csv')
plt.plot(df['PREGNANT'], df['DIABETES'], label='covid data')
plt.xlabel('PREGNANT')
plt.ylabel('PNEUMONIA')
plt.legend(loc="upper left")
plt.title('Scatter graph for covid data', color='blue')
plt.show()

#print(df.isnull().sum())

df.head()
```
