```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('covid.csv')

print(df.isnull().sum())
df.info()
print(df.describe())
```

```python
#line graph

plt.plot(df['MEDICAL_UNIT'], df['AGE'], label='covid data')
plt.xlabel('MEDICAL_UNIT',color='red')
plt.ylabel('AGE', color='orange')
plt.legend(loc="upper right")
plt.title('line graph for covid data', color='orange')
plt.show()

print(df['MEDICAL_UNIT'],df['AGE'])

```

```python
# Calculate mean values for each attribute
mean_pneumonia = df['PNEUMONIA'].mean()
mean_diabetes = df['DIABETES'].mean()

# Create bar plotorange
plt.bar(['PNEUMONIA', 'DIABETES'], [mean_pneumonia, mean_diabetes],label="Attributes")

# Add labels and title
plt.xlabel('Attributes', color= "blue")
plt.ylabel('Mean Values',  color= "red")
plt.title('Bar plot in COVID Data', color="Orange")
plt.legend(loc='upper right')

# Show plot
plt.show()

print("Mean mean_pneumonia is : ",mean_pneumonia)
print("The mean_diabetes is:", mean_diabetes)
```

```python
# Create violin plot
plt.violinplot([df['PNEUMONIA'], df['AGE']])

# Add labels and title
plt.xlabel('PNEUMONIA',color="orange")
plt.ylabel('AGE',color="blue")
plt.title('Violin Plot for PNEUMONIA and AGE in COVID Data' ,color="Red")

# Add legend
plt.legend(['PNEUMONIA', 'AGE'])

# Show plot
plt.show()
print(df['PNEUMONIA'], df['AGE'])
```
