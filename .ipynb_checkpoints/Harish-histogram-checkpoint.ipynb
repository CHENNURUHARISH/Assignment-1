```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Read data
df = pd.read_csv('covid.csv')

# Define bins for age groups
bins = np.arange(0, 110, 10)  # Age groups from 0 to 100+ in intervals of 10 years

# Create histogram plot
plt.hist(df['AGE'], bins=bins, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('PNEUMONIA')
plt.title('Histogram of Age Distribution for COVID Data')

# Show plot
plt.show()
```
