```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Read data
df = pd.read_csv('covid.csv')

# Create box plot
plt.boxplot([df['PNEUMONIA'], df['DIABETES']])

# Add labels and title
plt.xlabel('PNEUMONIA')
plt.ylabel('DIABETES')
plt.title('Box Plot for PNEUMONIA and DIABETES in COVID Data')

# Add legend
plt.legend(['PNEUMONIA', 'DIABETES'])

# Show plot
plt.show()

# Print first few rows of DataFrame
print(df.head())
```
