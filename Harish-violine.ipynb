```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Read data
df = pd.read_csv('covid.csv')

# Create violin plot
plt.violinplot([df['PNEUMONIA'], df['AGE']])

# Add labels and title
plt.xlabel('PNEUMONIA')
plt.ylabel('AGE')
plt.title('Violin Plot for PNEUMONIA and AGE in COVID Data')

# Add legend
plt.legend(['PNEUMONIA', 'AGE'])

# Show plot
plt.show()

# Print first few rows of DataFrame
print(df.head())
```
